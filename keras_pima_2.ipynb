```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("pima-indians-diabetes.data.csv",header = None,delimiter=',')
print(len(dataset))
dataset.head()
```

```python
df_labels = dataset.iloc[:,8] #labels
df_features = dataset.iloc[:,0:8] #features
df_features.isnull().sum() # to check null values
```

```python
print(df_features.shape)
#print(np.array(labels).shape)
```

```python
X_train,X_test,y_train,y_test =  train_test_split(df_features,df_labels,test_size=0.2)
print(X_train.shape)
print(X_test.shape)
#print(y_train.shape)
#print(y_test.shape)
np.array(y_train).shape
```

```python
model = Sequential()
model.add(Dense(500,input_dim = 8,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=1000,batch_size=80,validation_data=(X_test,y_test))
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```
